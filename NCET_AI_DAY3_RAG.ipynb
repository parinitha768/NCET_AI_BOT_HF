```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.litellm import LiteLLM
from llama_index.core.llms import ChatMessage
from llama_index.embeddings.huggingface import HuggingFaceEmbedding  # Add this import

# 1. Configure Embedding Model (Local Hugging Face)
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

# 2. Configure LLM (Groq LLAMA)
llm = LiteLLM("groq/llama-3.3-70b-versatile")

# 3. Load data with local embeddings
data = SimpleDirectoryReader(input_dir="./data/").load_data()
index = VectorStoreIndex.from_documents(
    data,
    embed_model=embed_model  # Critical: Use local embeddings
)

# 4. Create query engine with Groq LLM
query_engine = index.as_query_engine(llm=llm)  # Now works

# 5. Querying: Retrieve relevant chunks and generate response
response = query_engine.query("What is the main finding of the PDF?")
print(response)
```

```python
chat_engine = index.as_chat_engine(chat_mode="best", llm=llm, verbose=True)
chat_engine
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core import Settings

llm = LiteLLM("groq/llama-3.3-70b-versatile", temperature=0.7)
Settings.llm = llm
data = SimpleDirectoryReader(input_dir="./data/").load_data()
index = VectorStoreIndex.from_documents(data, settings=Settings,embed_model=embed_model )
query_engine = index.as_query_engine()
response = query_engine.query("What is Fractional Investment Does?")
print(response)
```

```python
response = chat_engine.stream_chat("How LLM transforming Tech World?")
for token in response.response_gen:
    print(token, end="")
```
